```python
# Sample test data for sales volumes of different products
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales_volumes = [150, 200, 120, 300, 250]

# Display the data
for product, sales in zip(products, sales_volumes):
    print(f"{product}: {sales}")


```

```python
%pip install matplotlib
import matplotlib.pyplot as plt

# Plotting the bar chart
plt.bar(products, sales_volumes, color='skyblue')
plt.xlabel('Products')
plt.ylabel('Sales Volumes')
plt.title('Sales Volumes of Products')
plt.show()
```
